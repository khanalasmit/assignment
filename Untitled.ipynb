```python
import numpy as np
import matplotlib.pyplot as plt

def get_user_input():
    func_str = input("Enter a function in terms of x (e.g., x**3 - x - 2): ")
    a = float(input("Enter lower guess a: "))
    b = float(input("Enter upper guess b: "))
    tol = float(input("Enter tolerance: "))
    return func_str, a, b, tol

def f(x, func_str):
    return eval(func_str, {"x": x, "np": np})

def bisection_method(func_str, a, b, tol):
    if f(a, func_str) * f(b, func_str) >= 0:
        print("Invalid interval. f(a) and f(b) must have opposite signs.")
        return None, []

    iterations = []
    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        iterations.append(c)
        f_c = f(c, func_str)

        if abs(f_c) < tol:
            break

        if f(a, func_str) * f_c < 0:
            b = c
        else:
            a = c

    return c, iterations

def plot_function(func_str, a, b, root, iterations):
    x = np.linspace(a - 1, b + 1, 400)
    y = [f(val, func_str) for val in x]

    plt.figure(figsize=(6,4))
    plt.axhline(0, color='gray', linestyle='--')
    plt.plot(x, y, label=f"f(x) = {func_str}")
    plt.plot(root, f(root, func_str), 'ro', label=f"Root ≈ {root:.6f}")
    plt.plot(iterations, [f(val, func_str) for val in iterations], 'kx', label="Iterations")

    plt.title("Bisection Method Convergence")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

# ----------------------------
# Main Program
# ----------------------------

# Uncomment the block below for manual input
func_str, a, b, tol = get_user_input()
root, iterations = bisection_method(func_str, a, b, tol)
if root is not None:
    print(f"Approximate root: {root:.6f}")
    plot_function(func_str, a, b, root, iterations)
'''
# Example 1
func1 = "x**3 - x - 2"
a1, b1, tol1 = 1, 2, 1e-6
root1, iters1 = bisection_method(func1, a1, b1, tol1)
print(f"Root of {func1} ≈ {root1:.6f}")
plot_function(func1, a1, b1, root1, iters1)

# Example 2
func2 = "np.cos(x) - x"
a2, b2, tol2 = 0, 1, 1e-6
root2, iters2 = bisection_method(func2, a2, b2, tol2)
print(f"Root of {func2} ≈ {root2:.6f}")
plot_function(func2, a2, b2, root2, iters2)

# Example 3
func3 = "x**2 - 4"
a3, b3, tol3 = 0, 3, 1e-6
root3, iters3 = bisection_method(func3, a3, b3, tol3)
print(f"Root of {func3} ≈ {root3:.6f}")
plot_function(func3, a3, b3, root3, iters3)
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_user_input():
    func_str = input("Enter a function in terms of x (e.g., x**3 - x - 2): ")
    a = float(input("Enter first initial guess (x0): "))
    b = float(input("Enter second initial guess (x1): "))
    tol = float(input("Enter tolerance: "))
    return func_str, a, b, tol

def f(x, func_str):
    return eval(func_str, {"x": x, "np": np})

def secant_method(func_str, x0, x1, tol, max_iter=100):
    iterations = [x0, x1]
    
    for _ in range(max_iter):
        f_x0 = f(x0, func_str)
        f_x1 = f(x1, func_str)

        if abs(f_x1 - f_x0) < 1e-12:
            print("Denominator too small. Possible division by zero.")
            return None, iterations

        x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)
        iterations.append(x2)

        if abs(x2 - x1) < tol:
            return x2, iterations

        x0, x1 = x1, x2

    print("Maximum iterations reached without convergence.")
    return None, iterations

def plot_function(func_str, a, b, root, iterations):
    x = np.linspace(min(a, b) - 1, max(a, b) + 1, 400)
    y = [f(val, func_str) for val in x]

    plt.figure(figsize=(6,4))
    plt.axhline(0, color='gray', linestyle='--')
    plt.plot(x, y, label=f"f(x) = {func_str}")
    if root is not None:
        plt.plot(root, f(root, func_str), 'ro', label=f"Root ≈ {root:.6f}")
    plt.plot(iterations, [f(val, func_str) for val in iterations], 'kx', label="Iterations")

    plt.title("Secant Method Convergence")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

# ----------------------------
# Main Program
# ----------------------------

# Uncomment the block below for manual input
func_str, a, b, tol = get_user_input()
root, iterations = secant_method(func_str, a, b, tol)
if root is not None:
    print(f"Approximate root: {root:.6f}")
plot_function(func_str, a, b, root, iterations)

'''
# Example 1
func1 = "x**3 - x - 2"
a1, b1, tol1 = 1, 2, 1e-6
root1, iters1 = bisection_method(func1, a1, b1, tol1)
print(f"Root of {func1} ≈ {root1:.6f}")
plot_function(func1, a1, b1, root1, iters1)

# Example 2
func2 = "np.cos(x) - x"
a2, b2, tol2 = 0, 1, 1e-6
root2, iters2 = bisection_method(func2, a2, b2, tol2)
print(f"Root of {func2} ≈ {root2:.6f}")
plot_function(func2, a2, b2, root2, iters2)

# Example 3
func3 = "x**2 - 4"
a3, b3, tol3 = 0, 3, 1e-6
root3, iters3 = bisection_method(func3, a3, b3, tol3)
print(f"Root of {func3} ≈ {root3:.6f}")
plot_function(func3, a3, b3, root3, iters3)
'''
```
