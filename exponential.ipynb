```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

class ExponentailCurve:
    def __init__(self, x_values, y_values):
        self.x_vals = x_values
        self.y_vals = y_values
        self.a = None
        self.b = None

    def f(self, x):
        return self.a + self.b * x

    def evaluate(self):
        x = np.array(self.x_vals)
        y = np.array(np.log(self.y_vals))
        n = len(x)
        
        sum_x = np.sum(x)
        sum_y = np.sum(y)
        sum_x2 = np.sum(x ** 2)
        sum_xy = np.sum(x * y)

        D = n * sum_x2 - sum_x ** 2
        D_a = sum_y * sum_x2 - sum_x * sum_xy
        D_b = n * sum_xy - sum_x * sum_y

        self.a = D_a / D
        self.b = D_b / D
        self.a=np.exp(self.a)
        return self.a, self.b

    def plot(self):
        plt.scatter(self.x_vals, self.y_vals, color='blue', edgecolor='red')
        x = np.linspace(min(self.x_vals), max(self.x_vals), 100)
        y = [self.a*np.exp(self.b*i) for i in x]
        plt.plot(x, y, color='green')
        plt.grid(True)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Exponential Curve Fitting - Asmit Khanal')
        plt.show()
```

```python
x_values=[2,4,5,8,10,12]
y_values=[3,5,10,18,48,90]
```

```python
er=ExponentailCurve(x_values,y_values)
er.evaluate()
```

```python
er.plot()
```
