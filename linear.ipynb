```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

class LinearCurve:
    def __init__(self, x_values, y_values):
        self.x_vals = x_values
        self.y_vals = y_values
        self.a = None
        self.b = None

    def f(self, x):
        return self.a + self.b * x

    def evaluate(self):
        x = np.array(self.x_vals)
        y = np.array(self.y_vals)
        n = len(x)
        
        sum_x = np.sum(x)
        sum_y = np.sum(y)
        sum_x2 = np.sum(x ** 2)
        sum_xy = np.sum(x * y)

        D = n * sum_x2 - sum_x ** 2
        D_a = sum_y * sum_x2 - sum_x * sum_xy
        D_b = n * sum_xy - sum_x * sum_y

        self.a = D_a / D
        self.b = D_b / D
        return self.a, self.b

    def plot(self):
        plt.scatter(self.x_vals, self.y_vals, color='blue', edgecolor='black')
        x = np.linspace(min(self.x_vals), max(self.x_vals), 100)
        y = [self.a + self.b * xi for xi in x]
        plt.plot(x, y, color='green')
        plt.grid(True)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Linear Curve Fitting - Asmit Khanal')
        plt.show()
```

```python
x_values=[1,2,3,4,5]
y_values=[2.2,2.8,3.6,4.5,5.1]
```

```python
lr=LinearCurve(x_values,y_values)
print(lr.evaluate())
```

```python
lr.plot()
```
