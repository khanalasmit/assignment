```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class PolyFit:
    def __init__(self,xvalues,yvalues):
        self.xvalues=np.array(xvalues)
        self.yvalues=np.array(yvalues)
        self.m=3#degree of polynomial
        self.C=None
    def evaluate(self):
        A=np.zeros((self.m+1,self.m+1))
        B=np.array([0,0,0,0])
        for i in range(self.m+1):
            for j in range(self.m+1):
                A[i, j] = np.sum(self.xvalues ** (i + j))
            B[i]=np.sum(self.xvalues**(i)*self.yvalues)
        self.C=np.linalg.solve(A,B)
    def plot(self):
        self.evaluate()
        plt.scatter(self.xvalues,self.yvalues)
        X_range = np.linspace(min(self.xvalues), max(self.xvalues), 100)
        Y_range = [np.polyval(np.flip(self.C), x) for x in X_range]
        
        plt.plot(X_range, Y_range, color='green')
        plt.title('Polynomial fit of the curve - Asmit Khanal')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.grid(True)
        plt.show()

        
        
```

```python
p=PolyFit([1,2,3,4,5,6],[1,34,67,100,156,200])
```

```python
p.plot()
```
